# Matplotlib Practice

### Load Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

##### Create Sample Data

```python
days = [1,2,3,4,5,6,7]
temperature = [50,51,52,48,47,49,46]
```

##### Plot the Data using the matplotlib

```python
plt.plot(days, temperature, color='green', linewidth = 5)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Sample Plot')
plt.show()
```
