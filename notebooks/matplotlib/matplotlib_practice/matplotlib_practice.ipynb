# Matplotlib Practice

### Load Matplotlib

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
````

##### Create Sample Data

````python
days = [1,2,3,4,5,6,7]
temperature = [50,51,52,48,47,49,46]
````

##### Plot the Data using the matplotlib

````python
plt.plot(days, temperature, color='green', linewidth = 5)
plt.title('Temperature in Week')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.show()
````

##### Format Strings in plot Function
 Example is the string in the plot function contains 'g+'

````python
plt.plot(days, temperature, 'g+')
plt.title('Temperature in Week')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.show()
````

Look at the parameters inside the plot function

````python
plt.plot(days, temperature, color='green', marker='+', linestyle='', markersize=15)
plt.title('Temperature in Week')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.show()
````

##### Axes labels, legend, Grid

````python
# SAMPLE DATA
days = [1,2,3,4,5,6,7]
max_temperature = [50,51,52,48,47,49,46]
min_temperature = [43,42,40,44,33,35,37]
avg_temperature = [45,48,48,46,40,42,41]
````

````python
# PUT SOME LABELS
plt.title('Temperature in Week')
plt.xlabel('Days')
plt.ylabel('Temperature')

# PLOT THE DATA AND PUT SOME LABELS ON IT
plt.plot(days, max_temperature, label='Max')
plt.plot(days, min_temperature, label='Min')
plt.plot(days, avg_temperature, label='Avg')
plt.legend(loc='lower left')
plt.show()
````

##### Bar Chart

````python
company = ['GOOGLE', 'AMAZON', 'MSFT', 'FACEBOOK']
revenue = [90, 136, 89, 27]
profit = [40,2,34,12]
````

````python
x_position = np.arange(len(company))
x_position
````

Use 'xticks()' function to use the company name instead the value of y_position

````python
plt.xticks(x_position, company)
plt.bar(x_position, revenue, label='Revenue')
plt.bar(x_position, profit, label='Profit')
plt.title('US Teck Stocks')
plt.xlabel('Company Name')
plt.ylabel('Revenue(Billion)')
plt.legend()
plt.show()
````

##### Separated Charts

Some parts in these line separate the charts. Especially the subtraction and addition on it's position and setting the width of the charts.

```python
plt.bar(x_position-0.2, revenue, width=0.4, label='Revenue')
plt.bar(x_position+0.2, profit, width=0.4,  label='Profit')

````python
plt.xticks(x_position, company)
plt.bar(x_position-0.2, revenue, width=0.4, label='Revenue')
plt.bar(x_position+0.2, profit, width=0.4,  label='Profit')
plt.title('US Teck Stocks')
plt.xlabel('Company Name')
plt.ylabel('Revenue(Billion)')
plt.legend()
plt.show()
````

##### Can also use a horizontal charts

Use 'barh()' function to make the chart horizontal. Just change the name of the variables for more accurate information.

````python
plt.yticks(x_position, company)
plt.barh(x_position-0.2, revenue, label='Revenue')
plt.barh(x_position+0.2, profit,  label='Profit')
plt.title('US Teck Stocks')
plt.xlabel('Company Name')
plt.ylabel('Revenue(Billion)')
plt.legend()
plt.show()
````

### Histogram

````python
sugar_level = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
````

````python
plt.hist(sugar_level)
plt.show()
````

##### Use 'bins' parameter to set the bins of the histogram. the default bin is 10.

The maximum value of 'rwidth' parameter is 1.

````python
plt.hist(sugar_level, bins=3, rwidth=0.95)
plt.show()
````

##### Can also set a modified bins using list.

````python
plt.hist(sugar_level, bins=[80, 100, 125, 150], rwidth=0.95)
plt.show()
````

##### Modify the color of the chart using 'color' parameter.

````python
plt.hist(sugar_level, bins=3, rwidth=0.95, color='g')
plt.show()
````

##### Can also set the histogram type using 'histtype' parameter.

````python
plt.hist(sugar_level, bins=3, rwidth=0.95, color='g', histtype='step')
plt.show()
````

##### Plotting 2 types of Histograms

Sample Data

````python
sugar_level_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
sugar_level_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
````

Plotting

````python
plt.title('Blood Sugar Chart')
plt.xlabel('Sugar Level Range')
plt.ylabel('Number of Patients')
plt.hist([sugar_level_men, sugar_level_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['green', 'orange'], label=['Men', 'Women'])
plt.legend()
plt.show()
````

##### Horizontal Histogram

````python
plt.title('Blood Sugar Chart')
plt.ylabel('Sugar Level Range')
plt.xlabel('Number of Patients')
plt.hist([sugar_level_men, sugar_level_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['green', 'orange'], label=['Men', 'Women'], orientation='horizontal')
plt.legend()
plt.show()
````

### Pie Chart

Sample Data

````python
expenses = [1400, 600, 300, 410, 250]
labels = ['Home Rent', 'Food', 'Internet', 'Car', 'Others']
````

Plotting

````python
plt.axis('equal')
plt.pie(expenses, labels=labels, autopct='%0.2f%%')
plt.show()
````

Change the size of the pie chart using the argument 'radius'

````python
plt.axis('equal')
plt.pie(expenses, labels=labels, autopct='%0.2f%%' ,radius=1.5)
plt.show()
````

Explode Argument

````python
plt.axis('equal')
plt.pie(expenses, labels=labels, autopct='%0.2f%%' ,radius=1.5, explode=[0,0.2,0.2,0,0])
plt.show()
````

Change the angle using 'startangle' argument

````python
plt.axis('equal')
plt.pie(expenses, labels=labels, autopct='%0.2f%%' ,radius=1.5, explode=[0,0.2,0.2,0,0], startangle=180)
plt.show()
````
